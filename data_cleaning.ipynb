# Begin Data Cleaning

```python
movie=pd.read_csv('final_fetch_yeah.csv',index_col=0)
```

```python
Internet_movie = []
Rotten_tomatoes = []
Metacritic = []
count_I=0
count_R=0
count_M=0
```

```python
movie.Ratings.values[0]
```

```python
movie.Ratings.values[0]
```

```python
for i in range(len(movie.Ratings.values)):
    for dic in movie.Ratings.values[i]:
        if dic['Source']=='Internet Movie Database':
            score = dic['Value']
            score = float(score.split('/')[0].encode('ascii'))
            Internet_movie.append(score)
            count_I+=1
        if dic['Source']=='Rotten Tomatoes':
            score = dic['Value']
            score = float(score.split('%')[0].encode('ascii'))/10
            Rotten_tomatoes.append(score)
            count_R+=1
        if dic['Source']=='Metacritic':
            score = dic['Value']
            score = float(score.split('/')[0].encode('ascii'))/10
            Metacritic.append(score)
            count_M+=1
    if count_I < i+1:
        Internet_movie.append(np.nan)
        count_I+=1
    if count_R < i+1:
        Rotten_tomatoes.append(np.nan)
        count_R+=1
    if count_M < i+1:
        Metacritic.append(np.nan)
        count_M+=1
```

```python
movie['Internet_Movie_Database']=Internet_movie
movie['Rotten_Tomatoes']=Rotten_tomatoes
movie['Metacritic']=Metacritic
```

```python
movie.drop(['Country','DVD','Language','Poster','Ratings','Released','Response','Error','totalSeasons','Website','Type'],axis=1,inplace=True)
```

```python
movie.drop(['Rated'],axis=1,inplace=True)
```

```python
movie.reset_index(drop=True, inplace=True)
```

```python
movie.to_csv('data_selected_yeah.csv')
```

```python
movie=pd.read_csv('data_selected_yeah.csv',index_col=0)
movie
```

```python
def getDfSummary(movie):
    number_nan=movie.isnull().sum()
    
#     ads=ads.dropna()
    number_distinct=movie.nunique()

    result=pd.concat([number_nan,number_distinct,movie.describe()[1:].transpose()], axis=1)
    result.rename(columns={result.columns[0]:'number_nan',result.columns[1]:'number_distinct'}, inplace=True)

    return result
```
